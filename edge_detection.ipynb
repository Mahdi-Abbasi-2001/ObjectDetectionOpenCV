```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('data/puppy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
edges1 = cv2.Canny(img, threshold1=127, threshold2=127)
plt.imshow(edges1)
```

```python
med_val = np.median(img)
med_val
```

```python
lower = int(max(0, 0.7*med_val))
upper = int(min(255, 1.3*med_val))
```

```python
edges2 = cv2.Canny(img, threshold1=lower, threshold2=upper)
plt.imshow(edges2)
```

```python
blurred_img = cv2.blur(img, ksize=(5, 5))
```

```python
edges3 = cv2.Canny(blurred_img, threshold1=lower-10, threshold2=upper-100)
plt.imshow(edges3)
```
