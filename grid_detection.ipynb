```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('data/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
flat_chess_found, flat_chess_corners = cv2.findChessboardCorners(flat_chess, (7, 7))
```

```python
cv2.drawChessboardCorners(flat_chess, (7, 7), flat_chess_corners, flat_chess_found)
```

```python
plt.imshow(flat_chess)
```

```python
dots = cv2.imread('data/dot_grid.png')
dots = cv2.cvtColor(dots, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dots)
```

```python
dots_found, dots_corners = cv2.findCirclesGrid(dots, (10, 10), cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
cv2.drawChessboardCorners(dots, (10, 10), dots_corners, dots_found)
```

```python
plt.imshow(dots)
```
