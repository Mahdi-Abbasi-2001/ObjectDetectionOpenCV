```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('data/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
flat_chess_gray = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(flat_chess_gray, cmap="gray")
```

```python
flat_chess_corners = cv2.goodFeaturesToTrack(flat_chess_gray, 64, 0.01, 10)
```

```python
flat_chess_corners = np.intp(flat_chess_corners)
```

```python
for flat_chess_corner in flat_chess_corners:
    x, y = flat_chess_corner.ravel()
    cv2.circle(flat_chess, (x, y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(flat_chess)
```
