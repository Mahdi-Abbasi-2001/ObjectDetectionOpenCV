```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('data/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
flat_chess_gray = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(flat_chess_gray, cmap="gray")
```

```python
flat_chess_gray = np.float32(flat_chess_gray)
```

```python
flat_chess_res = cv2.cornerHarris(src=flat_chess_gray, blockSize=2, ksize=3, k=0.04)
```

```python
flat_chess_res = cv2.dilate(flat_chess_res, None)
```

```python
flat_chess[flat_chess_res > 0.01 * flat_chess_res.max()] = [255, 0, 0]
```

```python
plt.imshow(flat_chess)
```
